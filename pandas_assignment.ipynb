```python
import pandas as pd
```

```python
data = {
     'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
     'Age': [24, 27, 22, 32, 29],
     'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
     'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)
df.head()
df.to_csv('salary.csv', index=False)
```

```python
print("Summary Statistics:\n")
print(df[['Age','Salary']].describe())
avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("\nAverage HR Salary:", avg_salary)
```

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
filtr = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print(filtr)
```

```python
avg = df.groupby('Department')['Salary'].mean()
print(avg)
```

```python
sort = df.sort_values(by='Salary', ascending=True)
print(sort)
sort.to_csv('sorted_salary.csv', index=False)
```
