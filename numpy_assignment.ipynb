```python
import numpy as np
import random 
```

```python
arr = np.arange(1,21)
print(arr)
```

```python
summ = np.sum(arr)
meann = np.mean(arr)
mediann = np.median(arr)
std = np.std(arr)
ind = np.where(arr > 10)
print("Sum -",summ)
print("Mean -",meann)
print("Median -",mediann)
print("Standard Deviation -",std)
print("Indices of Numbers greater than 10 -",ind[0])
```

```python
arr1 = np.arange(1,17).reshape(4,4)
print("Array:\n", arr1)
transpose_arr = arr1.T
print("\nTranspose:\n", transpose_arr)
row_sum = np.sum(arr1, axis=1)
col_sum = np.sum(arr1, axis=0)
print("\nRow-wise sum:", row_sum)
print("Column-wise sum:", col_sum)
```

```python
arr2 = np.random.randint(1,21,(3,3))
arr3 = np.random.randint(1,21,(3,3))
print("Two 3X3 matrices -\n",arr2,"\n\n",arr3)
print()

add = arr2 + arr3
sub = arr2 - arr3
multi = arr3 * arr3
dot = np.dot(arr2,arr3)
print("Addition =\n",add)
print()
print("Subtraction =\n",sub)
print()
print("Multiplication =\n",multi)
print()
print("Dot Product =\n",dot)
```

```python
arr4 = np.arange(1, 13)
print("1D array :",arr4)
arr2D = arr4.reshape(3, 4)
print("\n3x4 Array:\n", arr2D)
sliced = arr2D[:2, -2:]
print("\nSliced Array (first 2 rows, last 2 columns):\n", sliced)
```
